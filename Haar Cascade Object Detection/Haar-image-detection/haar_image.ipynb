# Basic Object Detection

```python
import cv2
vehicle_cascade = cv2.CascadeClassifier(r"C:/Users/DANISH LAPTOP/Desktop/Projects/haar_image/cars.xml")
def detect_vehicles(image_path):

    image = cv2.imread(image_path)
    if image is None:
        print("Error: Image not found or path is incorrect.")
        return

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    vehicles = vehicle_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)

    for (x, y, w, h) in vehicles:
        cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

    cv2.imshow("Vehicle Detection", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

detect_vehicles(r"C:/Users/DANISH LAPTOP/Desktop/Projects/haar_image/66.jpeg")
detect_vehicles(r"C:/Users/DANISH LAPTOP/Desktop/Projects/haar_image/55.jpeg")
```
