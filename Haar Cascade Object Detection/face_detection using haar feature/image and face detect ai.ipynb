```python
import cv2

def detect_eyes(image_path):
    # Load the eye cascade
    eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

    # Read the image
    img = cv2.imread(image_path)

    if img is None:
        print("Error: Could not read the image.")
        return

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect eyes
    eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Draw rectangles around the detected eyes
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(img, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

    # Display the output
    cv2.imshow('Image with Eye Detection', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
image_path = 'C:/Users/DANISH LAPTOP/Desktop/PIC.jpg'  # Adjust the path accordingly
detect_eyes(image_path)
```

```python
import cv2

def detect_faces_and_eyes(image_path):
    # Load the face and eye cascades
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

    # Read the image
    img = cv2.imread(image_path)

    if img is None:
        print("Error: Could not read the image.")
        return

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    # Draw rectangles around detected faces and detect eyes within each face
    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)  # Draw face rectangle
        
        # Region of interest for eyes
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = img[y:y + h, x:x + w]
        
        # Detect eyes within the face region
        eyes = eye_cascade.detectMultiScale(roi_gray, scaleFactor=1.1, minNeighbors=5)
   # Draw rectangles around the detected eyes
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

# Display the output
    cv2.imshow('Image with Face and Eye Detection', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
image_path = 'C:/Users/DANISH LAPTOP/Desktop/c.jpeg'  
detect_faces_and_eyes(image_path)
```
